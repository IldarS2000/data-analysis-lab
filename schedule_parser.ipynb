```python
from selenium import webdriver
from bs4 import BeautifulSoup


def get_all_days_schedule(schedule):
    days = ['понедельник', 'вторник', 'среда', 'четверг', 'пятница', 'суббота']
    time = ['08:30-10:00', '10:10-11:40', '11:50-13:20', '14:00-15:30', '15:40-17:10', '17:50-19:20']
    for week_day, day_schedule in zip(days, schedule):
        print(f'день недели: {week_day}')
        for t, subject in zip(time, day_schedule):
            print(f'{t}: {subject}')
        print()


def get_specific_day_schedule(schedule, day_number):
    days = ['понедельник', 'вторник', 'среда', 'четверг', 'пятница', 'суббота']
    time = ['08:30-10:00', '10:10-11:40', '11:50-13:20', '14:00-15:30', '15:40-17:10', '17:50-19:20']
    print(f'день недели: {days[day_number - 1]}')
    for t, subject in zip(time, schedule[day_number - 1]):
        print(f'{t}: {subject}')
    print()


def main():
    group = input('введите вашу группу: ')
    schedule_choice = int(input('на какой день хотите посмотреть расписание\n'
                                'примечание: (введите цифру без точки)\n'
                                '0. все дни\n'
                                '1. понедельник\n'
                                '2. вторник\n'
                                '3. среда\n'
                                '4. четверг\n'
                                '5. пятница\n'
                                '6. суббота\n'
                                'ваш ввод: '))
    print()

    url = 'https://kpfu.ru/studentu/ucheba/raspisanie'
    driver = webdriver.Firefox(executable_path='geckodriver.exe')
    driver.get(url)

    element_enter = driver.find_element_by_xpath('//*[@id="p_group_name"]')
    element_enter.send_keys(group)

    button = driver.find_element_by_xpath('//*[@id="ss_content"]/div[2]/div/div/form/div/div[1]/div/div/div[3]')
    button.click()

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')

    table = soup.find_all('table')[-1]
    trs = table.find_all('tr')[1:-1]
    schedule = [[] for i in range(6)]
    for tr in trs:
        tds = tr.find_all('td')[1:]
        for i in range(6):
            spans = tds[i].find_all('span')
            if spans:
                if isinstance(spans, list):
                    schedule[i].append(', '.join([span.text for span in spans]))
                else:
                    schedule[i].append(spans.text)
            else:
                schedule[i].append('-')

    if schedule_choice == 0:
        get_all_days_schedule(schedule)
    else:
        get_specific_day_schedule(schedule, schedule_choice)

    driver.quit()


if __name__ == '__main__':
    main()
```
